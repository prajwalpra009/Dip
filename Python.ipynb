```python
#local scope variables and global variables
def spam():
    eggs=100 #local variable of spam
    print(eggs)
    
def chase():
    eggs=10 #local variable of chase
    ham=10
    print(ham)
    print(eggs)
    spam()

chase()
eggs='global'
print(eggs) #global

"""o/p : 10
         10
         100 
        global  """
```

```python
#global statement
def eggs():
    global cheese #global
    cheese='egg'
    egg='cheese'
    print(egg)
    print(cheese)
    
ham='eggs'
eggs()
print(cheese)

'''o/p: cheese
        egg
        egg  '''
```

```python
#exceptions

def spam(divideby):
    try:
        return 42/divideby
    except ZeroDivisionError:
        print('error:invalid')

print(spam(43))
print(spam(34))
print(spam(0))
print(spam(90))
# if try and except was put for print statements only first 3 o/p would be there as except doesnt loop back
```

```python
#collatz

def collatz(number):
    while number!=1:
        if number%2==0:
            number=number//2
            print(number)
        else:
            number=3*number+1
            print(number)
        
print('enter a number')
i=int(input())
collatz(i)
```

```python
#extra snippets
#prime number

def prime(m,n):
    for num in range(m,n+1):
        if num>1:
            for i in range(2,num):
                if num%i==0:
                    break
            else:
                print(num)
                    
print('enter range')
m=int(input())
n=int(input())
if m<0 or n<0:
    print('invalid')
    
prime(m,n)

#fibonaci
def fib(m):
    if m<0:
        print('invalid')
    elif m==0:
        print('1')
    elif m==2:
        return 1
    else:
        return fib(m-1)+fib(m-2)
    
print(fib(4))

#factorial
```

```python
#working with lists
catname=[]
while True:
    print('enter the names of your cat else enter nothing to exist')
    name=input()
   
    if name=='':
        break
    catname.append(name) #append

    
print(catname)

str='hello how are you' #string slicing
print(str[0:3])

#refrences
spam = [0, 1, 2, 3, 4, 5]
cheese = spam # The reference is being copied, not the list.
cheese[1] = 'Hello!' # This changes the list value.[0, 'Hello!', 2, 3, 4, 5]
# The cheese variable refers to the same list.[0, 'Hello!', 2, 3, 4, 5]
print(spam)
print(cheese)

#copy and deepcopy
import copy
spam=[2,3,44,5,[2,3]]
cheese=copy.copy(spam) #copy
cheese[1]=4
print(cheese)
print(spam)
ham=copy.deepcopy(spam)#deepcopy
ham[0]=3
print(ham)
print(spam)
```
